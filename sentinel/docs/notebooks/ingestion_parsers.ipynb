# How to read unstructured log files? 

## Built-in Methods for Reading Common Log Formats

This project includes predefined methods to read common log formats such as **WAS**, **HDC**, **HSM**, and **IBMMQ**.

#### Example: Reading WAS Logs

```python
from sentinel.ingestion.ingestor import LogIngestor

df_was = LogIngestor.ingest("data/WAS.log", "WAS")
df_was.head()
```

#### Example: Reading HDC Logs

```python
df_hdc = LogIngestor.ingest("data/HDC_1.log", "HDC")
df_hdc.head()
```

#### Example: Reading HSM Logs

```python
df_hsm = LogIngestor.ingest("data/HSM.log", "HSM")
df_hsm.head()
```

#### Example: Reading IBMMQ Logs

```python
df_ibm = LogIngestor.ingest("data/IBM_MQ.txt", "IBM_MQ")
df_ibm.head()
```
